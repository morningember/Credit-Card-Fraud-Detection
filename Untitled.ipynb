## Credit card Fraud detection

```python
#importing useful libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sample = pd.read_csv('creditcard.csv')
```

```python
sample.head(5)
```

```python
print(f'Data set has {sample.shape[0]} rows and {sample.shape[1]} column')
```

```python
sample.info()
```

```python
# Find missing Values
sample.isna().sum()
```

### Finding and dropping duplicates

```python
sample.duplicated().any()
```

```python
sample = sample.drop_duplicates()
```

```python
print(f'Data set has {sample.shape[0]} rows and {sample.shape[1]} column')
```

```python
#Drop time as it is unimportant
sample = sample.drop(['Time'], axis=1)
```

```python
# Statistical information
sample.describe()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sample.Amount = sc.fit_transform(pd.DataFrame(sample.Amount))
```

```python
sample.head()
```

```python
sns.countplot(sample.Class)
```

### it could be seen that the data is unbalanced. 
### To balance the data, select same amount of data points with [class = 0] as there is in [class = 1]

##### transactions with [class = 0] => normal trasaction and [class = 1] => fraudulent trasaction

### Splitting Data into normal transactions and fraudulent transactions

```python
nml_trans = sample[sample.Class == 0]
fraud_trans = sample[sample.Class == 1]
```

```python
print(nml_trans.shape)
print(fraud_trans.shape)
```

#### Sampling the normal transactions to equal fraudulent transactions (Under Sampling)

```python
nml_trans_sampled = nml_trans.sample(n=473)
```

```python
nml_trans_sampled.shape
```

```python
# Merging sampled normals transaction and fraudulent transactions 
data = pd.concat([nml_trans_sampled, fraud_trans], axis=0)
```

```python
data.shape
```

```python
sample = sample.drop('Class', axis=1)
row = sample.iloc[0]
```

```python
row
```

```python
data.head()
```

```python
#Finding Correlation between features and target
corr = data.corr()
plt.figure(figsize=(40, 20))
sns.heatmap(corr, annot=True)
```

### Splitting Data 

```python
X = data.drop('Class', axis=1)
y = data.Class
X.iloc[0]
```

```python
row = X.iloc[5].values
```

```python
#importing Useful libraries
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify=y,random_state=2)
```

### Importing and using machine learning algorithms to train data

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(X_train, y_train)
```

```python
pred = log.predict(X_test)
```

```python
log.predict(row.values.reshape(1,-1))
```

### Finding the accuracy of the prediction of the test samples

```python
from sklearn.metrics import accuracy_score
print('Accuracy of logistic regression is: ', accuracy_score(pred, y_test))
```
